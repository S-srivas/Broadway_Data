```python
# link: http://www.broadwayleague.com/research/statistics-broadway-nyc/
# Importation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
from pylab import bar, show, plot, scatter
import seaborn as sns
import timeit
```

```python
# Reading the main file
df=pd.read_csv('broadway.csv', parse_dates=['Year'], index_col='Year')
```

```python
# Plotting the Attendance column
def plot_df_2plots(df, x, y, title="", xlabel='Date', ylabel='Attendance in Millions', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.bar(x, y)
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=90)
    plt.show()

plot_df_2plots(df, x=df.index, y=df.Attendance, title='Broadway Attendance Over the Years')
```

```python
# In this plot we're plotting the Gross column
def plot_df_plot(df, x, y, title="", xlabel='Year', ylabel='Gross Accumulation in Millions', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.bar(x, y, color='purple')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=90)
    plt.show()

plot_df_plot(df, x=df.index, y=df.Gross, title='Money Accumulated Over the Years')
```

```python
# In this plot we're plotting the Playign weeks column
def plot_df(df, x, y, title="", xlabel='Year', ylabel='# of Weeks Played', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.bar(x, y, color='green')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=90)
    plt.show()

plot_df(df, x=df.index, y=df.Playing_weeks, title='# Played Over the Years')
```

```python
# In this plot we're plotting the # of new productions
def plot_(df, x, y, title="", xlabel='Year', ylabel='# of New Productions', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.bar(x, y, color='red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=90)
    plt.show()

plot_(df, x=df.index, y=df.New_Productions, title='# of New Productions Over the Years')
```

```python
# Plotting the Attendance and Gross
# Reading the file w/ different parameters

df=pd.read_csv('broadway.csv', parse_dates=['Year'], index_col='Attendance')

def lot_df(df, x, y, title="", xlabel='Attendance in Millions', ylabel='Gross in Millions', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.scatter(x, y)
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=90)
    plt.show()

lot_df(df, x=df.index, y=df.Gross, title='Broadway Attendance V. Gross Accumulation')
```
